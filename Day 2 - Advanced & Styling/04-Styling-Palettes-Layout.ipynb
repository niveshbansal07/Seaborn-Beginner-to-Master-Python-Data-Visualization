#### Styling, Palettes & Layout

1) Global Styling with sns.set_theme()

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid", context="talk", palette="deep")
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
ax = sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")

ax.set_title("Tips vs Total Bill")     
ax.set_xlabel("Total Bill ($)")        
ax.set_ylabel("Tip ($)")               

sns.despine()  # Remove top/right spines (the borders around the plot)
plt.tight_layout()
plt.show()
```
