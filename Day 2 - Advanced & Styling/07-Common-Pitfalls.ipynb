#### Common Pitfalls

1) Overplotting (too many points)
- Problem: Scatterplots get messy if many points overlap.
- **Solution**:
- Reduce transparency â†’ alpha=0.5
- Use kdeplot or hexbin (Matplotlib)

````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# 1. Reduce Opacity(alpha)
sns.scatterplot(data=tips, x="total_bill", y="tip", alpha=0.5)
plt.show()

# 2. Use KDE plot
sns.kdeplot(data=tips, x="total_bill", y="tip", fill=True, cmap="Reds")
plt.show()

# 3. Use Hexbin plot
plt.hexbin(tips["total_bill"], tips["tip"], gridsize=20, cmap="Blues")
plt.colorbar(label="Count")
plt.show()
````

2) Default labels are not enough

3) Comparing groups
- **Problem**: Donâ€™t just use barplot mean values â†’ hides distribution & outliers.
- **Solution**: Use boxplot, violinplot, or stripplot alongside barplot.

````python
sns.violinplot(data=tips, x="day", y="tip", hue="sex", split=True)
sns.stripplot(data=tips, x="day", y="tip", hue="sex", dodge=True, alpha=0.5)
````

4) Consistent color semantics
- **Problem**: Using different colors for the same group across charts â†’ confusing.
- **Solution**: Pick one palette for a category and stick to it.

````python
sns.set_theme(palette="muted")  # use same palette across charts
````

**Quick Checklist**
- Check for overplotting â†’ use alpha/kde/hexbin.
- Always set titles, axes, legends.
- Use distribution plots for group comparisons.
- Keep colors consistent.
