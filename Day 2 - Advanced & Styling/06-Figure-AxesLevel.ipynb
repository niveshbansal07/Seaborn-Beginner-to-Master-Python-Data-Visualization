#### Figure-level vs Axes-level

1) Axes-level functions

- **Examples**:
scatterplot, lineplot, histplot, kdeplot, boxplot, violinplot, heatmap, regplot
- These return a Matplotlib Axes object (ax).
- You control subploting manually with plt.subplots().
- Good when you want precise control.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
fig, ax = plt.subplots()

sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax)
ax.set_title("Scatterplot (Axes-level)")
plt.show()
```

2) Figure-level functions

- **Examples**:
relplot, catplot, jointplot, pairplot, lmplot
- These create a full figure (can include multiple subplots).
- Layout & facets handled automatically.
- Good for quick exploration.

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="sex")
plt.show()
```

| Feature  | Axes-level functions                     | Figure-level functions                        |
| -------- | ---------------------------------------- | --------------------------------------------- |
| Examples | scatterplot, lineplot, heatmap, regplot… | relplot, catplot, jointplot, pairplot, lmplot |
| Returns  | Axes object (`ax`)                       | Figure/Grid object                            |
| Subplots | You manage manually with Matplotlib      | Automatic grid/faceting                       |
| Use case | Fine control, custom plots               | Quick exploration, multiple facets            |


Quick Analogy

- Axes-level → like drawing one painting on a canvas yourself.
- Figure-level → like asking Seaborn to arrange a gallery of paintings for you.
