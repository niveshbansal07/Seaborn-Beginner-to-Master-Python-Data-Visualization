#### Faceting & Small Multiples
- Split data into small subplots based on a category.

1) FacetGrid (low level)

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

smoker = sns.FacetGrid(tips, col="smoker")
sex = sns.FacetGrid(tips, col="sex")
smoker.map_dataframe(sns.scatterplot, x="total_bill", y="tip", hue="sex")
smoker.add_legend()
sex.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
plt.show()
```

- One scatterplot for Smokers, one for Non-smokers.

2. relplot (Shortcut for scatter/line)

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="sex", s=150)
plt.show()
```

- Split into Lunch vs Dinner. Colors = male/female.

3. catplot (shortcut for categories)

```python
sns.catplot(data=tips, x="total_bill", y="tip", col="sex", kind="box")
plt.show()
```

- Boxplots of tip across days, split by sex.
