#### Correlations & Heatmaps

What is correlation?

Correlation measures how two numeric variables move together.

- Range = -1 to +1.
- +1 → perfect positive (both increase together).
- -1 → perfect negative (one goes up, other goes down).
- 0 → no relation.

1. Correlation matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips[["total_bill", "tip", "size"]].corr()
print(corr)
```

- Gives table of pairwise correlations.

2. Heatmap (easy to read)

```python

corr = tips[["total_bill", "tip", "size"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.show()
```

- You can quickly “see” strong vs weak relationships.

3. Clustered Heatmap

```python
corr = tips[["total_bill", "tip", "size"]].corr()
sns.clustermap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

- total_bill ↔ tip → strong positive (bigger bill → bigger tip).
- size ↔ total_bill → positive (larger table → bigger bill).
- tip ↔ size → weaker, but still positive.
