#### Legends, Annotations & Saving

- A good chart must explain itself â†’ labels, legend, notes.
- Save at high DPI (300+) for clear reports.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

ax = sns.regplot(data=tips, x="total_bill", y="tip")
ax.annotate(
    "Higher tips with higher bills",
    color="red",
    xy=(40,7),    #point to mark (x=40, y=7)
    xytext=(25,8.5),  # text position
    arrowprops=dict(arrowstyle="->", color="red")
)

# Remove legend if not needed
if ax.legend_:
    ax.legend_.remove()

plt.tight_layout()
plt.savefig("tips_scatter.png", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
```
