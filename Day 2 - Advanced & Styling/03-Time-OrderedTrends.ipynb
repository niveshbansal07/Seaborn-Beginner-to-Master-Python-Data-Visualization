#### Time/Ordered Trends

- Sometimes we have ordered variables → like time (day, month, year) or size (1,2,3…).
For these, we don’t just scatter, we show a trend line.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
avg = tips.groupby("size", as_index=False)["tip"].mean()
sns.lineplot(data=avg, x="size", y="tip", marker="o")
plt.title("Average Tip by Party Size")
plt.show()
```

- Each dot = average tip for that table size.

```python
avg2 = tips.groupby("day", as_index=False)["total_bill"].mean()
sns.lineplot(data=avg2, x="day", y="total_bill", marker="o")
plt.title("Average Bill by Day")
plt.show()
```

- Days are ordered (Thu → Fri → Sat → Sun).

- Larger party size = higher average tip.
- Bills are highest on weekends (Sat/Sun).
