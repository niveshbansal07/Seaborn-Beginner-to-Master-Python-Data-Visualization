### Point Plot

- Shows mean per category as points with a line connecting categories; includes confidence intervals.
- Show trend across ordered categories (e.g., Thur→Fri→Sat→Sun).
- Compare groups with less ink than bars.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(6,3))
sns.pointplot(x="day", y="tip", data=tips)
plt.title("Average Tip Trend by Day (Point Plot)")
plt.show()

# With hue to compare groups
plt.figure(figsize=(6,3))
sns.pointplot(x="day", y="tip", hue="sex", data=tips, dodge=True)
plt.title("Average Tip by Day and Sex (Point Plot)")
plt.show()
```

- Point height = mean; vertical line = uncertainty.
- Line helps see upward/downward trend across categories.
