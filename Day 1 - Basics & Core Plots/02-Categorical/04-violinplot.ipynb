### Violin Plot

- Box plot + KDE (density shape). Shows full distribution per category.
- Need shape + summary together.
- Compare distributions across categories with more detail than box plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(6,3))
sns.violinplot(x="day", y="tip", data=tips, inner="box")  # inner="box"/"quartile"/None
plt.title("Tip Distribution by Day (Violin)")
plt.show()

# With split by a second category
plt.figure(figsize=(6,3))
sns.violinplot(x="day", y="tip", hue="sex", data=tips, split=True)
plt.title("Tip Distribution by Day (Split by Sex)")
plt.show()
```

- Thick areas = many values (high density).
- Thin areas = fewer values
