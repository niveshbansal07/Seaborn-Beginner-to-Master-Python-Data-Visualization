### Box Plot

- Shows distribution summary (median, quartiles) + outliers per category.
- Compare spread & outliers across categories.
- Robust summary for reports/dashboards.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(6,3))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Distribution by Day (Box Plot)")
plt.show()

# Add raw points for context
plt.figure(figsize=(6,3))
sns.boxplot(x="day", y="total_bill", data=tips, showcaps=True)
sns.stripplot(x="day", y="total_bill", data=tips, color="black", alpha=0.4)
plt.title("Box + Points")
plt.show()
```
