#### Countplot

- A bar plot where each bar shows the count (frequency) of observations for each category.
**OR**
- Bar chart showing how many times each category appears in your data(Shows how many rows are in each category).

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure(figsize=(5,3))
sns.countplot(x="day", data=tips)
plt.title("Count by Day")
plt.show()

# With sub-categories (hue)
plt.figure(figsize=(6,3))
sns.countplot(x="day", hue="sex", data=tips)
plt.title("Count by Day and Sex")
plt.show()
```

- Taller bar - more rows in that category.
- Hue shows split inside each bar.
