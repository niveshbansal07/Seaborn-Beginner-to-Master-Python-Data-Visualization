#### Bar Plot

- Shows the mean (or other aggregate) of a numeric value for each category.
- USE --
- Average tip per day/sex/â€¦
- Compare groups on a metric.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(6,3))
sns.barplot(x="day", y="tip", data=tips)  # default = mean
plt.title("Average Tip by Day")
plt.show()

# Grouped (compare inside each day)
plt.figure(figsize=(6,3))
sns.barplot(x="day", y="tip", hue="sex", data=tips, errorbar=None)
plt.title("Average Tip by Day and Sex")
plt.show()

# Use a different aggregate (e.g., sum)
from numpy import sum as np_sum
plt.figure(figsize=(6,3))
sns.barplot(x="day", y="tip", data=tips, estimator=np_sum, errorbar=None)
plt.title("Total Tip by Day (Sum)")
plt.show()
```

- Bar height = average (or chosen aggregate).
- Error lines = uncertainty (95% CI by default).
