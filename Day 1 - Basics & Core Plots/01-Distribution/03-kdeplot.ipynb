#### kdeplot

- `kdeplot` is Seaborn’s **direct function** for drawing the **Kernel Density Estimate (KDE)** curve.  
- Unlike `histplot`, it **doesn’t show bars (histogram)** by default → it only shows the smooth curve.  

1) Basic KDE

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
plt.show()
```

2) Filled KDE

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.show()
```

3) Compare Groups with hue

```python
sns.kdeplot(data=tips, x="total_bill", hue='sex', fill=True, alpha=0.5)
plt.show()
```

4) 2D KDE (for two variables)

```python
sns.kdeplot(data=tips, x="total_bill", y="tip", fill=True, cmap="mako")
plt.show()
```

```python
# Show all colormap names
print(plt.colormaps())
```

✅ Where KDE makes sense

Histogram → Yes (KDE is often added on top).

Rug Plot / Strip Plot → Yes (KDE can overlay density).

Scatter Plot → Yes (2D KDE shows density of points).

Pair Plot → Yes (can use kind="kde").

Joint Plot → Yes (can use kind="kde").

Violin Plot → Already uses KDE inside by default.
