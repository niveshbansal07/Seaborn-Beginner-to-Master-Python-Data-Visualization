### Seaborn Basics: Axes-level vs Figure-level

#### 1) Axes-level functions
- Create just **one small plot** on a specific Axes.  
- You can control *where* the plot goes.  
- Very flexible, works well with Matplotlib.  
- **Examples:** `sns.scatterplot()`, `sns.histplot()`, `sns.boxplot()`

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

#### 2) Figure-level functions

- Automatically create a new figure.
- Often make multiple plots in a grid layout.
- Easy to control layout with col=, row=, etc.
- **Examples:** `sns.catplot(), sns.pairplot(), sns.displot()`

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="sex")
plt.show()
```

One-line summary

Axes-level = Small plot inside an axes.

Figure-level = Big plot that manages its own figure and layout.
