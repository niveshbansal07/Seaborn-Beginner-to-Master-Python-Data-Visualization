#### Box Plot

A Box Plot is a statistical graph that shows the distribution of a numerical variable using its five-number summary:
- Minimum
- 1st Quartile (Q1 → 25%)
- Median (Q2 → 50%)
- 3rd Quartile (Q3 → 75%)
- Maximum

It also shows outliers separately.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

sns.boxplot(x=tips["total_bill"])
plt.show()
```
