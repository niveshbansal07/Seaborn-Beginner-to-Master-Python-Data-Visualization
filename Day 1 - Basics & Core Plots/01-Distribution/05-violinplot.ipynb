#### Violin Plot

- A Distribution Plot (used to show the spread and shape of numerical data).
- It combines:
- KDE curve (smooth density) → shows distribution shape.
- Box Plot → shows median, quartiles, and outliers.
- The shape looks like a violin, hence the name Violin Plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

sns.violinplot(x=tips["total_bill"])
plt.show()
```
