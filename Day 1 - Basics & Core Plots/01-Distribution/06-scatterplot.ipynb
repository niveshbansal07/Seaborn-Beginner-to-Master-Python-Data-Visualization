#### Scatter plot
- Scatter plot is distribution of two variables.
- A scatter plot shows data as dots (points) on a graph.
- Each point has two values → one on the x-axis and one on the y-axis.
- It helps us see if there’s a relationship (pattern, trend, or correlation) between two numerical variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()


# Not category-based by default, but categories can be added using hue.

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")

plt.show()
```
