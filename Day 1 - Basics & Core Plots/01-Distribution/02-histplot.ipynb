#### histplot - histogram

- A histogram shows **how data is distributed**.  
- It divides data into **bins (intervals)** and counts how many values fall inside each bin.  
- Itâ€™s great for seeing the **shape of your data** (like normal, skewed, uniform).

1. Basic Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

sns.histplot(data=tips, x="total_bill")
plt.show()
```


2) Histogram with KDE

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

3) Grouped Histogram with hue

```python
sns.histplot(data=tips, x="total_bill", hue="sex", kde=True, bins=20)
plt.show()
```
